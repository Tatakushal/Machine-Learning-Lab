```python
import numpy as np
dt = np.dtype([('Roll_No','i4'),('Name','U10'),('Marks','i4')])

data = [
    (101,'Ravi',85),
    (102,'Priya',92),
    (103,'Kiran',78),
    (104,'Anu',88),
    (105,'Sita',95),
]

array = np.array(data,dtype=dt)
print(array)
print(array[:3])
avarage_marks = np.mean(array['Marks'])
print(avarage_marks)
highest_marks = np.max(array['Marks'])
print(highest_marks)
lowest_marks = np.min(array['Marks'])
print(lowest_marks)

marks = array['Marks']
conditions = [
    array['Marks']>=90,
    (array['Marks']>=80) & (array['Marks']<=89),
    array['Marks']<80
]
choices = ['A','B','C']
grades = np.select(conditions,choices)
array_grades = np.append((array,grades))
print(array_grades)
```
