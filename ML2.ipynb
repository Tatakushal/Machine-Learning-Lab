```python
import numpy as np
import pandas as pd

df = pd.read_csv('Student.csv')

print(df)
print(df.head())
print(df.tail()) 
print(df.shape)
print(df.columns)
print(df.dtypes)
df.info()
```

```python
Attendance_mean = df['Attendance'].mean()
print(Attendance_mean)
Average_marks_Internal = df['InternalMarks'].mean() 
print(Average_marks_Internal)
Average_marks_External = df['ExternalMarks'].mean() 
print(Average_marks_External)
Max_marks_internal = df['InternalMarks'].max()
print(Max_marks_internal)
Min_marks_internal = df['InternalMarks'].min()
print(Min_marks_internal)
Max_marks_external = df['ExternalMarks'].max()
print(Max_marks_external)
Min_marks_external = df['ExternalMarks'].min()
print(Min_marks_external)
```

```python
df['TotalMarks'] = df['InternalMarks'] + df['ExternalMarks']
df['Percentage'] = df['TotalMarks']/100 * 100
print(df)
```

```python
high_attendance_students = df[df['Attendance'] > 90]
print(high_attendance_students)
high_totalmarks_students = df[df['TotalMarks'] > 90]
print(high_totalmarks_students)
female_students = df[df['Gender'] == 'F']
print(female_students)
male_students = df[df['Gender'] == 'M']
print(male_students)
low_externalmarks_students = df[df['ExternalMarks'] < 50]
print(low_externalmarks_students)
```

```python
highest_scoring_student = df.loc[df['TotalMarks'].idxmax()]
print(highest_scoring_student)
lowest_scoring_student = df.loc[df['TotalMarks'].idxmin()]
print(lowest_scoring_student)
top_three = df.nlargest(3, 'TotalMarks')
print(top_three)
last_three = df.nsmallest(3, 'TotalMarks')
print(last_three)
average_marks_total = df['TotalMarks'].mean()
print(average_marks_total)
df[df['TotalMarks'] > average_marks_total]
```

```python
male_average = df[df['Gender'] == 'M']['TotalMarks'].mean()
print(male_average)
female_average = df[df['Gender'] == 'F']['TotalMarks'].mean()
print(female_average)
attendance_comparison = df.groupby('Gender')['Attendance'].mean()
print(attendance_comparison)
```

```python
import matplotlib.pyplot as plt
df.set_index('Name')['TotalMarks'].plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Marks of All Students', fontsize=14)
plt.xlabel('Student Name', fontsize=12)
plt.ylabel('Total Marks', fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
df['Attendance'].plot(kind='hist', color='lightgreen', edgecolor='black')
plt.title('Distribution of Student Attendance', fontsize=14)
plt.xlabel('Attendance (%)', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
gender_counts = df['Gender'].value_counts()
gender_counts.plot(
    kind='pie', 
    startangle=90,    
    colors=['lightblue', 'pink'],
    wedgeprops={'edgecolor': 'black', 'linewidth': 1} 
)
plt.title('Gender Distribution of Students', fontsize=14)
plt.ylabel('') 
plt.show()
```

```python
import matplotlib.pyplot as plt
df.plot(kind='scatter', x='Attendance', y='TotalMarks', color='purple', edgecolor='black')
plt.title('Relationship Between Attendance and Total Marks', fontsize=14)
plt.xlabel('Attendance (%)', fontsize=12)
plt.ylabel('Total Marks', fontsize=12)
plt.show()
```

```python
correlation = df['Attendance'].corr(df['TotalMarks'])
print(correlation)
```
